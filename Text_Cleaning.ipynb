```python
import numpy as np
import csv
csv_filename = '/content/MalayStopWord_1.csv'
with open(csv_filename) as f:
    reader = csv.reader(f)
    stopwords = list(reader)
    #print(stopwords)



#import required modules
from functools import reduce
from operator import concat

#reducing the list
flat=reduce(concat,stopwords)
#displaying the flattening of the list
print("the reduced list is=",flat)
  

```

```python
flat
```

```python
import numpy as np
import re

#stopwords = ["for", "on", "an", "a", "of", "and", "in", "the", "to", "from"]

def clean_text(tweet):
    if type(tweet) == np.float:
        return ""
    temp = tweet.lower()
    temp = re.sub("'", "", temp) # to avoid removing contractions in english
    temp = re.sub("@[A-Za-z0-9_]+","", temp)
    temp = re.sub("#[A-Za-z0-9_]+","", temp) 
    temp = re.sub(r'http\S+', '', temp)
    temp = re.sub('[()!?]', ' ', temp)
    temp = re.sub('\[.*?\]',' ', temp)
    temp = re.sub("[^a-z0-9]"," ", temp)
    temp = re.sub(r'[0-9]', '', temp)
    temp = temp.split()
    temp = [w for w in temp if not w in flat]
    temp = " ".join(word for word in temp)
    return temp
```

```python
results = [clean_text(tw) for tw in text]
```
